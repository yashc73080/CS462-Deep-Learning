```python
c = torch.arange(0,30)
print(c)
```

```python
d = torch.randn((2,3)) # size of tensor for normal distribution
print(d)
```

```python
# Tensor with shape (2,3,4)
e = torch.randn((2,3,4))
print(e)
```

```python
# Tensor with shape (2,3,4) with numbers from 1 to 24
f = torch.arange(1,25).reshape(2,3,4)
print(f)
```

```python
# Add 1 to every item in tensor -> broadcast
c + 1
```

```python
a = torch.arange(0,20)
a + c
```

```python
a * 2
```

```python
# Element-wise multiplication
a = torch.arange(0,20)
b = torch.arange(0,20)
a * b
```

```python
# Matrix multiplication
a = torch.randn((3,2))
b = torch.randn((3,2))
print(a)
print(b)
print(b.T) # Transpose to make the dimensions match
a @ b.T
```

```python
# Indexing and slicing

a = torch.randn((1000, 3, 100, 100)) # ex: 1000 images, 3 channels, 100x100 pixels each
a[0, 0, 0, 0] # First image, R channel, top left corner
```

```python
a[:10, :, 0, 0] # top left corner from top 10 images (all channels for each)
```

```python
a[:10, :, 0, 0].shape
```

```python
a[10:, 0, :10, :10] # R channel for last 10 images, top left 100 pixels
```

```python
# Want to change the images to start with images, put channel at the end
b = a.permute((0, 2, 3, 1)) # original dim at id 0 stays, original dim of 1 moves to the end, etc
b
```

```python
b.shape
```

```python
c = a.transpose(1,3)
c.shape
```
